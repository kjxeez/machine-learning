```python
df['MonthlyAverage'] = df['MonthlyAverage'].astype(float)
df['MonthlyAveragePDKss'] = df['MonthlyAveragePDKss'].astype(float)
df['MonthlyAverage']
```

Вычислю средние значения параметров по округам

```python
mean_df = df.groupby(['AdmArea', 'Parameter']).agg({
    'MonthlyAverage': 'mean',
    'MonthlyAveragePDKss': 'mean' 
}).reset_index()
mean_df
```

```python
mean_df[mean_df['AdmArea']=='Зеленоградский административный округ']
```

```python
import plotly.express as px
import pandas as pd

# Создание DataFrame из предоставленных данных


# Создание графика в Plotly
fig = px.bar(mean_df, x='AdmArea', y='MonthlyAverage', color='Parameter', title='Средние значения параметров по округам')

# Добавление интерактивности
fig.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args=[{'y': [mean_df['MonthlyAverage']]}],
                    label='Среднее значение MonthlyAverage',
                    method='update'
                ),
                dict(
                    args=[{'y': [mean_df['MonthlyAveragePDKss']]}],
                    label='Среднее значение MonthlyAveragePDKss',
                    method='update'
                )
            ]),
            direction="down",
            showactive=True,
            xanchor="left",
            yanchor="top"
        )
    ]
)
```
